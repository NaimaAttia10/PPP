```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torch.optim.lr_scheduler import StepLR
from torch.utils.data import DataLoader
import numpy as np
import torchvision
from torchvision import datasets, models
from torchvision import transforms


def train_model(model, criterion, optimizer, scheduler, n_epochs = 5):
    losses = []
    acc = []
    test_accuracies = []
    
    # set the model to train mode initially
    model.train()
    for epoch in range(n_epochs):
        running_loss = 0.0
        running_corrects = 0.0
        for i, data in enumerate(train_loader, 0):

            # get the inputs and assign them to cuda
            inputs, labels = data
            #inputs = inputs.to(device).half() # uncomment for half precision model
            inputs = inputs.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            
            # forward + backward + optimize
            outputs = model(inputs)
            _, preds = torch.max(outputs.data, 1)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            # calculate the loss_acc 
            running_loss += loss.item()/inputs.size(0)
            running_corrects += ((labels==preds).sum().item())/inputs.size(0)
        
        epoch_loss = running_loss/len(train_loader)
        epoch_accuracy = running_corrects/len(train_loader)
        print("Epoch %s,  loss: %.4f, accurracy: %.4f" % (epoch+1, epoch_loss, epoch_accuracy))
        
        losses.append(epoch_loss)
        acc.append(epoch_accuracy)
        
        # switch the model to eval mode to evaluate on test data
        model.eval()
        test_acc = eval_model(model)
        test_accuracies.append(test_acc)
        
        # re-set the model to train mode (after validation)
        model.train()
        scheduler.step(test_acc)
    print('Finished Training')
    return model, losses, acc, test_accuracies

def eval_model(model):
    corrects = 0.0
    total = 0.0
    with torch.no_grad():
        for i, data in enumerate(val_loader, 0):
            images, labels = data
            images = images.to(device)
            labels = labels.to(device)
            
            outputs = model(images)
            _, preds = torch.max(outputs.data, 1)
            
            total += labels.size(0)
            corrects += (preds == labels).sum().item()

    test_acc = 100.0 * corrects / total
    print('Accuracy of the network on the test images: %d %%' % (test_acc))
    return test_acc

if __name__ == '__main__':

        train_transforms = transforms.Compose([
        transforms.RandomResizedCrop(244),
        transforms.RandomHorizontalFlip(p=0.5),
        transforms.RandomRotation(35),
        transforms.RandomAdjustSharpness(sharpness_factor=2, p=0.5),
        transforms.RandomGrayscale(p=0.5),
        transforms.RandomPerspective(distortion_scale=0.5, p=0.5),
        transforms.RandomPosterize(bits=2, p=0.5),
        transforms.ToTensor(),
        transforms.Normalize(
            [0.485, 0.456, 0.406],
            [0.229, 0.224, 0.225]
        )
    ])

    val_transforms = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            [0.485, 0.456, 0.406],
            [0.229, 0.224, 0.225]
        )
    ])

    train_dataset = datasets.ImageFolder(
        root='/kaggle/input/cards-image-datasetclassification/train',
        transform=train_transforms
    )
    val_dataset = datasets.ImageFolder(
        root='/kaggle/input/cards-image-datasetclassification/valid',
        transform=val_transforms
    )

    train_loader = DataLoader(
        train_dataset,
        batch_size=4,
        shuffle=True,
        num_workers=2
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=4,
        shuffle=True,
        num_workers=2
    )

    model = models.efficientnet_b1(pretrained=True)
    for params in model.parameters():
            params.requires_grad = True
    model.classifier[1] = nn.Linear(in_features=1280, out_features=53) 


    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"Using '{device}' to train the Model !")

    model = model.to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)
    lrscheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', patience=3, threshold = 0.9)

    model, training_losses, training_accuracy, test_accuracy = train_model(model, criterion, optimizer, lrscheduler, n_epochs=12)
    
    
    torch.save(model.state_dict(), "./trainedModelFinal.pt")
    
    f, tab = plt.subplots(2,2, figsize = (10, 8))
    tab[0, 0].plot(training_losses)
    tab[0, 0].set_title("Training loss")
    tab[0, 1].plot(training_accuracy)
    tab[0, 1].set_title("Training accuracy")
    tab[1, 0].plot(test_accuracy)
    tab[1, 0].set_title("Test accuracy")
    
    
    model.eval()
```
